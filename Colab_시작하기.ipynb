```
import pandas as pd
cnts = pd.read_csv('read-counts.txt', sep='\t', comment='#', index_col=0)
cnts.head()
```

```
cnts['clip_enrichment'] = cnts['CLIP-35L33G.bam'] / cnts['RNA-control.bam']
cnts['rden_change'] = (cnts['RPF-siLin28a.bam'] / cnts['RNA-siLin28a.bam']) / (cnts['RPF-siLuc.bam'] / cnts['RNA-siLuc.bam'])
cnts.head()
```

```
from matplotlib import pyplot as plt
import numpy as np

fig, ax = plt.subplots(1, 1, figsize=(5, 5))
ax.scatter(np.log2(cnts['clip_enrichment']),
           np.log2(cnts['rden_change']))
```

```
import matplotlib.pyplot as plt
import numpy as np

# Filter out NaN or infinite values in 'clip_enrichment' and 'rden_change' before log2 transformation
# This addresses the RuntimeWarning encountered previously.
plot_data = cnts.replace([np.inf, -np.inf], np.nan).dropna(subset=['clip_enrichment', 'rden_change'])

# Apply log2 transformation
log2_clip_enrichment = np.log2(plot_data['clip_enrichment'])
log2_rden_change = np.log2(plot_data['rden_change'])

fig, ax = plt.subplots(1, 1, figsize=(8, 6))

ax.scatter(log2_clip_enrichment,
           log2_rden_change, alpha=0.6, s=10)

ax.set_title('Log2 Enrichment vs. Log2 Ribosome Density Change (Figure 4d-like)')
ax.set_xlabel('log2(CLIP Enrichment)')
ax.set_ylabel('log2(Ribosome Density Change)')

ax.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
